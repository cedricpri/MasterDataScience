# Python

```python
import random
import matplotlib.pyplot as plt
```

```python
# use a for loop to generate a walk
position = [0, 0] #Initial position x, y
walk = [] #List in which we will keep the walk obtained
steps = 10000

for i in range(steps):
    random1 = random.randint(0,3) #Gives the direction of the movement (0 is to the left, 1 to the top, 2 to the right, 3 to the bottom)

    if random1 == 0:
        movement = [-1, 0]
    elif random1 == 1:
        movement = [0, 1]
    elif random1 == 2:
        movement = [1, 0]
    else:
        movement = [0, -1]
    
    newPosition = [position[0] + movement[0], position[1] + movement[1]]
    walk.append(newPosition)
    position = newPosition
```

```python
#print(walk)
x,y=[], []
for index, value in enumerate(walk): #Needed for the scatter function?
    x.append(value[0])
    y.append(value[1])
plt.plot(x, y)
```

# Numpy

```python
import numpy as np
```

```python
n_stories = 10000
t_max = 200
```

```python
#We consider that we can only move -1 or 1 -> if not, we can just change np.random.choice to np.random.int by choosing the parameter
steps = [np.random.choice([-1, 1], size=n_stories), np.random.choice([-1, 1], size=n_stories)] #Direction of the movement
positions = np.cumsum(steps, axis=1)
plt.plot(positions[0], positions[1])
```

```python
np.unique(steps) # Verification, must print 'array([-1,  1])'
```

## Print max and min positions

```python
print(positions.max())
print(positions.min())
```

## Walk > 30

```python
toCheck = positions>30
np.any(toCheck, axis=1)
```

## Crossing times

```python
print(np.argmax(positions>30))
```
